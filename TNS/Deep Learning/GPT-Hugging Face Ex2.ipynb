```python
# Step 1: Install transformers (if not already installed)
# !pip install transformers

# Step 2: Import the pipeline
from transformers import pipeline

# Step 3: Load the Q&A pipeline
qa = pipeline("question-answering")

# Step 4: Provide a context (paragraph) and a question
context = """
Once, a rabbit mocked a slow tortoise and challenged him to a race.
Confident of his speed, the rabbit quickly raced ahead and, seeing the tortoise far behind,
decided to rest under a tree. Certain of victory, he fell asleep.
Meanwhile, the tortoise kept moving slowly but steadily without stopping.
Step by step, he passed the sleeping rabbit and continued toward the finish line.
When the rabbit finally woke up, he ran as fast as he could, but it was too late.
The tortoise had already crossed the finish line. Moral: Slow and steady wins the race.
"""

question = "What is the moral of the story"

# Step 5: Get the answer
answer = qa(question=question, context=context)

# Step 6: Print the result
print("Question:", question)
print("Answer:", answer["answer"])
```
